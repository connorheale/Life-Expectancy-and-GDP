# Introduction

The objective of this project is to investigate GDP data from the World Bank and life expectancy data from the World Health Organization. The goal is first examine the variation in GDP and life expectancy among different countries and then explore possible connections between the two with different plots to visualize the relationship.


**Sources:**

- GDP Source: [World Bank](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD) national accounts data, and OECD National Accounts data files.

- Life expectancy Data Source: [World Health Organization](http://apps.who.int/gho/data/node.main.688)

## Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Loading & Examining the Data

The next step is to read the csv file with pandas and examine the columns as well as the first few rows

```python
df = pd.read_csv('all_data.csv')
df.head()
```

I will use .info() to ensure all Dtypes are correct according to their value type.

```python
df.info()
```

Because .head() only shows the first few rows, I will examine unique values that were not displayed with .head() to gain more information about the data.

```python
print(df['Year'].unique())
print(df['Country'].unique())
```

## Data Tidying

I will adjust the following column: 'Life expectancy at birth (years)', and rename it to the abbreviation 'LEAB' with .rename to simplify further examination with plots.

```python
df.rename(columns={'Life expectancy at birth (years)':'LEAB'}, inplace=True)
```

## Bar Plots

The next step is to accomplish the goal of examining variation between countries across GDP and LEAB. I chose to do this with individual bar plots for both GDP and LEAB by country

```python
labels = ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe']
sns.set_style('darkgrid')
plt.figure(figsize=(6,6))
g = sns.barplot(x='GDP', y='Country', data=df, palette='deep')
plt.xlabel('GDP in Trillions (USD)')
g.set(yticklabels=labels)
```

The United States GDP is far higher than the rest, leaving countries with lower GDP like Chile and especially Zimbabwe to be poorly respresented. I will separate by country in future plots to resolve this.

```python
plt.figure(figsize=(6,6))
g = sns.barplot(x='LEAB', y='Country', data=df, palette='deep')
plt.xlabel('Life Expectancy at Birth (Years)')
g.set(yticklabels=labels)
```

Life expectancy has much less variation among the different countries, though Zimbabwe still lags behind.

## Line Plots

I will use line plots to compare the growth and decline of GDP and LEAB over time

```python
plt.figure(figsize=(6,6))
sns.lineplot(x='Year', y='GDP', hue='Country', data=df, palette='deep', legend=False)
plt.legend(labels)
```

China and the USA had the most steady increase in GDP, while Chile and Zimbabwe's growth appears to be stagnant when compared the far greater GDP of other countries.

```python
plt.figure(figsize=(6,6))
sns.lineplot(x='Year', y='LEAB', hue='Country', data=df, palette='deep', legend=False)
plt.ylabel('Life Expectancy at Birth (Years)')
plt.legend(labels)
```

Zimbabwe had the most significant increase in life expectancy, and all other countries displayed appear to have had an increase in life expectancy as well.

## Separating by Country

The next step is to separate the growth over time by each country to resolve the issue of countries with lower values of GDP and life expectancy being outweighed by countries with greater values.

```python
gdp_graph = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', sharey=False, height=4)
gdp_graph.map(sns.lineplot, 'Year','GDP') \
    .add_legend()
```

The individual charts reveal that all countries had significant increases in GDP despite Zimbabwe and Chile showing little increase in previous plots.

```python
gdp_graph = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', sharey=False, height=4)
gdp_graph.map(sns.lineplot, 'Year','LEAB') \
    .add_legend()
```

Each country also had its life expectancy increase over time, with shapes that are similar to that of their GDP growth, respectively.

## Conclusion

This project's intent was to examine the relationship between life expetnacy and GDP across different countries. While initially some countries were overshadowed by countries with higher life expectancy and GDP, further investigation revealed that each country had an increase in both over time. Additionally, the peaks and valleys in the GDP line plots with unqiue y values seem to reflect that of the life expectancy plots that also had separate y values. This may suggest a relationship between GDP and life expectancy.
